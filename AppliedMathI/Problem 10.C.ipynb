Here we plot the solution of the equation for several values of c.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
t = np.linspace(0,4*np.pi) # The time
def xt(t,c):
    return (np.sin(t)-np.cos(t))/2 + c*np.exp(t)
xt_vec = np.vectorize(xt)
```

```python
plt.figure(1,figsize=(20,5))
plt.subplot(1,2,1)
plt.scatter(t,xt_vec(t,1.0),label="c=1",marker="*",s=100)
plt.scatter(t,xt_vec(t,0.5),label="c=0.5",marker=",",s=100)
plt.scatter(t,xt_vec(t,0.25),label="c=0.125",marker="|",s=100)
plt.scatter(t,xt_vec(t,0.125),label="c=0.125",marker="^",s=100)
plt.scatter(t,xt_vec(t,0.0),label="c=0.0",marker="4",s=100)
plt.legend()
plt.title("Solution versus Time")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.xlim([0,4*np.pi])
plt.ylim([-2.0,2.0])
plt.grid()

plt.subplot(1,2,2)
plt.scatter(t,xt_vec(t,1.0),label="c=1",marker="*",s=60)
plt.scatter(t,xt_vec(t,0.5),label="c=0.5",marker=",",s=60)
plt.scatter(t,xt_vec(t,0.25),label="c=0.125",marker="|",s=60)
plt.scatter(t,xt_vec(t,0.125),label="c=0.125",marker="^",s=60)
plt.scatter(t,xt_vec(t,0.0),label="c=0.0",marker="4",s=60)
plt.legend()
plt.title("Solution versus Time")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.savefig("Problem10.c.png",dpi=400)
```
