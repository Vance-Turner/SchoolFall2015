We change variables: $x_1=x$ and $x_2=y$ to maximize confusion.

Thus we seek to maximize: $f(x,y)=2x+5y$

Subject to:

$0\leq x \leq 4$

$0\leq y\leq 6$

$x+y\leq 8\Longrightarrow y\leq 8-x$

We have level lines at: $y=c-2/5x$

Thus, our problem is constrained to the first quadrant of the xy plane:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
xpts = np.linspace(0,10)
con1_xpts = np.array([4.,4.,4.,4.,4.,4.])
con1 = np.arange(-20,20,7)
con2 = xpts - xpts + 6.
con3 = 8.-xpts
```

```python
# The various level lines:
lev1 = 2.-2/5.*xpts
lev2 = 3.-2/5.*xpts
lev3 = 4.-2/5.*xpts
lev4 = 5.-2/5.*xpts
lev5 = 6.-2/5.*xpts
lev6 = 7.-2/5.*xpts
lev7 = 8.-2/5.*xpts
```

```python
plt.plot(con1_xpts,con1,label='$0\leq x\leq 4$')
plt.plot(xpts,con2,label='$0\leq y \leq 6$')
plt.plot(xpts,con3,label='$y\leq 8-x$')
plt.plot(xpts,lev1,'*',label='2-2/5x')
plt.plot(xpts,lev2,'^',label='3-2/5x')
plt.plot(xpts,lev3,'2',label='4-2/5x')
plt.plot(xpts,lev4,'3',label='5-2/5x')
plt.plot(xpts,lev5,'4',label='6-2/5x')
plt.plot(xpts,lev6,'o',label='7-2/5x')
plt.plot(xpts,lev7,'x',label='8-2/5x')
plt.legend(loc=2)
plt.xlim([0,5])
plt.ylim([0,7])
plt.grid()
```

We observe the maximum must be (2,6).

And $f(2,6)=2*2+5*6=34$
