**Problem 6.31**

Find the best fit to $f(x)=x$ on $0<x<1$ of the form:

$$\psi(x)=\sum_{n=1}^N c_n\sin(n\pi x)$$

Thus we seek to solve the following linear system:

$$\sum_{j=1}^{N} c_j<\phi_i,\phi_j>=<\phi_i,f>$$

which is:

$$\sum_{j=1}^{N} c_j<\sin(i\pi x),\sin(j\pi x)>=<\sin(i \pi x),x>$$

We are working in the space: $L^2(0,1)$ thus the inner product becomes:

$$\sum_{j=1}^{N} c_j\int_0^1\sin(i\pi x)\sin(j\pi x)dx=\int_0^1 \sin(i \pi x)xdx$$

Let's work on the left integral:

$\int\sin(i\pi x)\sin(j\pi x)dx$

Using integration by parts: $u=\sin(i\pi x) \rightarrow du=i\pi\cos(i\pi x)dx$

$dv/dx = \sin(j\pi x) \rightarrow v = -1/(j\pi)\cos(j\pi x)$

$\int\sin(i\pi x)\sin(j \pi x)dx=-1/(j\pi)\cos(j\pi x)\sin(i\pi x)+i/j\int\cos(i\pi x)\cos(j\pi x)dx$

Do integration by parts again:

$u=\cos(i\pi x)\rightarrow du = -i\pi\sin(i\pi x)dx$

$dv/dx = \cos(j\pi x) \rightarrow v=1/(j\pi)\sin(j\pi x)$

$\int \cos(i\pi x)\cos(j \pi x)dx=1/(j\pi)\sin(j\pi x)\cos(i\pi x) - \int 1/(j\pi)\sin(j \pi x)*-i\pi\sin(i\pi x)dx$

$=1/(j\pi)\sin(j\pi x)\cos(i\pi x)+i/j\int\sin(i\pi x)\sin(j \pi x)dx$

Thus we have:

$\int \sin(i\pi x)\sin(j\pi x)dx=-1/(j\pi)\cos(j\pi x)\sin(i\pi x)+i/(j^2\pi)\sin(j\pi x)\cos(i\pi x)+i^2/j^2\int \sin(i\pi x)\sin(j\pi x)dx$

$(1-i^2/j^2)\int \sin(i\pi x)\sin(j \pi x)dx=-1/(j\pi)\cos(j\pi x)\sin(i\pi x)+i/(j^2\pi)\sin(j\pi x)\cos(i\pi x)$

$\int \sin(i\pi x)\sin(j\pi x)dx=1/(1-i^2/j^2)\big[-1/(j\pi)\cos(j\pi x)\sin(i\pi x)+i/(j^2\pi)\sin(j\pi x)\cos(i\pi x)\big]$

$\int_0^1 \sin(i\pi x)\sin(j \pi x)dx=1/(1-i^2/j^2)\big[-1/(j\pi)\cos(j\pi 1)\sin(i\pi 1)+i/(j^2\pi)\sin(j\pi 1)\cos(i\pi 1)\big] - 1/(1-i^2/j^2)\big[-1/(j\pi)\cos(j\pi 0)\sin(i\pi 0)+i/(j^2\pi)\sin(j\pi 0)\cos(i\pi 0)\big]$

$=1/(1-i^2/j^2)[-1/(j\pi)\cos(j\pi)\sin(i\pi)+i/(j^2\pi)\sin(j\pi)\cos(i\pi)]$

And since $m,n$ are integers the integral equals zero:

$=0$

Thus, they are mutually orthogonal, obviously. We calculate when $i=j$:

$\int_0^1 \sin^2(i\pi x)dx=\big[\frac{x}{2} - \frac{\sin(i\pi x)\cos(i\pi x)}{2i\pi}]\big|_0^1=1/2$ where we used a CAS to perform the integration. (We could have used a trigonometric identity after substitution). We also invoke the fact that $i$ is an integer.

Thus we are now attempting to find:

$1/2c_i=\int_0^1 \sin(i\pi x)xdx=\big[\sin(i\pi x)/(i^2\pi^2) - x\cos(i\pi x)/(i\pi)]\big|_0^1$

$=\sin(i\pi 1)/(i^2\pi^2) - 1\cos(i\pi 1)/(i\pi)-[\sin(i\pi 0)/(i^2\pi^2) - 0\cos(i\pi 0)/(i\pi)]$

$=0-\cos(i\pi)/(i\pi)=-\cos(i\pi)/(i\pi)$

Thus we find that:

$c_i=-2\cos(i\pi)/(i\pi)$

Let's test this fit out (with n=5) numerically for some visual confirmation:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,1,1000)
yTrue = 1.*x
```

```python
def psi(xi,n=30):
    value = 0.
    for i in xrange(1,n):
        value += -2.*np.cos(i*np.pi)/(i*np.pi)*np.sin(i*np.pi*xi)
    return value
psiVect = np.vectorize(psi,excluded=('n'))
yFit1 = psiVect(x,n=10)
yFit2 = psiVect(x,n=50)
yFit3 = psiVect(x,n=100)
yFit4 = psiVect(x,n=200)
```

```python
plt.figure(4,figsize=(20,20))
plt.subplot(221)
plt.plot(x,yTrue,label='y=x')
plt.plot(x,yFit1,'*',label='Fit, N=10')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc=2)
plt.subplot(222)
plt.plot(x,yTrue,label='y=x')
plt.plot(x,yFit2,'*',label='Fit, N=50')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc=2)
plt.subplot(223)
plt.plot(x,yTrue,label='y=x')
plt.plot(x,yFit3,'*',label='Fit, N=100')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc=2)
plt.subplot(224)
plt.plot(x,yTrue,label='y=x')
plt.plot(x,yFit3,'*',label='Fit, N=200')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc=2)
```

As we can see the fit is quite good, except at the end point. This is common though when fitting with sinusoidal functions.
